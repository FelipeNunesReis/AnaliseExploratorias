```python
import pandas as pd
```

```python
aluguel_casa= pd.read_excel('/content/houses_rent.xlsx')
```

```python
aluguel_casa.head(5)
```

```python
aluguel_casa = aluguel_casa.rename(columns={'parking spaces':'parking_space','rent amount':'rent_amount','property tax':'property_tax','fire insurance':'fire_insurance'})
```

```python
aluguel_casa.head()
```

```python
aluguel_casa.dtypes
```

```python
aluguel_casa = aluguel_casa.astype({'rent_amount':'float','property_tax':'float','fire_insurance':'float','total':'float'})
```

```python
aluguel_casa.dtypes
```

```python
aluguel_casa.head(5)
```

```python
aluguel_casa.info()
```

```python
aluguel_casa[['area','rooms','bathroom','parking_space','rent_amount','property_tax','fire_insurance','total']].describe()
```

```python
aluguel_casa= aluguel_casa[(aluguel_casa['total']<=4000)]
aluguel_casa.head()
```

```python
aluguel_casa=aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept')]
aluguel_casa.head()
```

```python
aluguel_casa=aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept') & (aluguel_casa['rooms']==2) | (aluguel_casa['rooms']==3)]
aluguel_casa.head()
```

```python
aluguel_casa= aluguel_casa.query("total <= 4000 and animal == 'accept'")
aluguel_casa.head()
```

```python
aluguel_casa= aluguel_casa.query("total <= 4000 and animal == 'accept' and rooms == 2 or rooms ==3")
aluguel_casa.head()
```

```python
agg_city_price = aluguel_casa.groupby(['city'])['total']. agg('mean').reset_index()
agg_city_price.head()
```

```python
agg_bath_room_price = aluguel_casa.groupby(['bathroom','rooms'])['total'].agg('mean').reset_index()
agg_bath_room_price.head()
```

```python
import seaborn as sns
```

```python
ax = sns.barplot(data= agg_bath_room_price, x= 'rooms', y = 'total');
ax.set(xlabel = ' qtd quartos', ylabel = 'mdia de preos');
```

```python
ax = sns.barplot(data= agg_bath_room_price, x= 'bathroom', y = 'total', color= "#FF6B6B");
ax.set(xlabel = ' qtd banheiros', ylabel = 'mdia de preos');
```
