```python
import pandas as pd
import seaborn as sns
```

```python
venda = pd.read_excel('/content/varejo.xlsx')
venda.head()
```

```python
venda.info()
```

```python
venda.groupby('idcanalvenda').idcompra.nunique()
```

```python
venda['idcanalvenda'] = venda['idcanalvenda'].str.replace('APP', 'Aplicativo')
venda.groupby('idcanalvenda').idcompra.nunique()
```

```python
venda.groupby('Nome_Departamento').idcompra.nunique()
```

```python
venda.groupby('estado').idcompra.nunique()
```

```python
venda['Nome_Departamento'] = venda['Nome_Departamento'].str.replace(' ', '_')
venda.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas_sem_nulo = venda.dropna()
vendas_sem_nulo.info()
```

```python
venda['estado'].fillna("MS", inplace = True)
```

```python
venda.info()
```

```python
media_preco = venda['Preço'].mean()
venda ['Preço'].fillna(media_preco, inplace = True)
```

```python
media_preco
```

```python
venda.info()
```

```python
venda.head()
```

```python
venda.describe()
```

```python
venda[['Preço','Preço_com_frete']].describe()
```

```python
preco_errado = venda.query('Preço > Preço_com_frete')

vendas_correto = venda.query('Preço < Preço_com_frete')
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending = False).reset_index(),2)
```

```python
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending = False)
```

```python
vendas_correto['mes'] = vendas_correto['Data'].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby('mes').idcompra.nunique().sort_values(ascending = False)
```

```python
clientes = pd.read_excel('cliente_varejo.xlsx')
clientes.head()
```

```python
clientes[['idade','renda']].describe()
```

```python
clientes.info()
```

```python
clientes = clientes.astype({'renda':'float'})
clientes.info()
```

```python
vendas_cliente = vendas_correto.merge(clientes, how = 'left', on = 'cliente_Log')
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending = False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending = False).reset_index(),2)
agg_idade_bandeira.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,8))

plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color = 'coral')

plt.ylabel('Média de idade')

plt.title('Idade média por Bandeira', loc = 'center', fontsize = 18, fontweight = 'bold', color = 'gray');
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color = 'coral')

plt.ylabel('Renda média')

plt.title('Renda média por canal de venda', fontsize = 18, color = 'gray' );
```

```python
venda_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize = (15,5))

plt.xlabel('Data de venda')
plt.ylabel('Quantidade de vendas')
x=venda_por_data['Data']
y = venda_por_data['idcompra']

plt.plot(x, y, color = 'coral', linewidth = '2');
```

```python
agg_dept_preco = round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending = False).reset_index(),2)
agg_dept_preco.head()
```

```python
fig = plt.figure(figsize = (25,7))

plt.bar(agg_dept_preco['Nome_Departamento'], agg_dept_preco['Preço_com_frete'], color = 'coral')

plt.ylabel('Preço médio')

plt.title('Preço médio por departamento', loc= 'center', fontsize = 18, fontweight = 'bold', color = 'black');
```
